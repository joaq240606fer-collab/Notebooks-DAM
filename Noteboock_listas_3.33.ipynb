IN [4]: En este caso creamos la variable (L) la cual es un objeto mutable es decir su valor puede cambiar etsa variable contiene una lista de numeros osea que apunta hacia un objeto que es esa lista de numeros que pasa con la siguiente linea de codigo [M = L] le estamos pidiendo que copie el contenido de la variable (L) dentro de la variable (N) provocando que esta ultima apunte hacia el objeto lista y con esta siguiente linea de codigo [ L[0]= 24 ] modifica el primer numero de la linea y lo cambia por 24 osea pasa a apountar a un nuevo objeto que contiene la lista con el modificador mientras que M sigue apuntando a la anterior por eso al printar (L) aparece la lista modificada y al printar (M) la lista inicial.

```python
L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change

print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

IN [5]: En este caso creamos la variable (L) la cual es un objeto mutable es decir su valor puede cambiar etsa variable contiene una lista de numeros osea que apunta hacia un objeto que es esa lista de numeros que pasa con la siguiente linea de codigo [M = L[:]] le estamos pidiendo que copie el primer caracter de cada uno de los numeros de la lista en  la variable (L) dentro de la variable (N) provocando que esta ultima apunte hacia el objeto lista y con esta siguiente linea de codigo [ L[0]= 24 ] modifica el primer numero de la linea y lo cambia por 24 osea pasa a apountar a un nuevo objeto que contiene la lista con el modificador mientras que M sigue apuntando a la anterior por eso al printar (L) aparece la lista modificada y al printar (M) la lista inicial.

```python
L = [2, 3, 4]
M = L[:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed
```

IN [6]: En este caso creamos una variable (L) que apuntaa un objeto tipo lista y luego creamos otra vareiable (M) a la que le damos el valor de (L) es decir apuntan al mismo objeto el operador (==) compara el contenido de la lista mientras que el operador (is) compara el objeto al que apunta por eso en ambos casos printa true

```python
L = [1, 2, 3]
M = L 			    # M and L reference the same object

L == M 			    # True: Same values => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)

True

```

IN [8]:En este caso como el operador (==) compara el contenido de la lista mientras que el operador (is) compara el objeto al que apunta como en este caso a cada variable le porporcionamos valores de lista aunque las lsitas tengan el mismo contenido apuntan a diferentes objetos por eso el operador (==) resuelve con true y el operador (in) false.

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M and L reference different objects

print(
    "L == M",
    L == M 			# True: Same values
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```
