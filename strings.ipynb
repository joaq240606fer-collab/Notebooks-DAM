## Type-Specific Methods ##

El operador find sirve para buscar segmentos del valor y nos devuelve el numero de posicion de los mismos 

```python
S = 'SpamEggsSpam'
S.find('pa')
```

En este caso le indicamos que empiece la busqueda desde el indice de inicio 3

```python
S.find('pa', 3)
```

Tambien podemos buscar el valor de una variable 

```python
S.find(S)
```

Si no existe el caracter devuelve el valor -1

```python
S.find('z')
```

El operador replace sirve para remplazar segmento del valor 

```python
S.replace('Eggs', 'Bacon')
```

El operador upper sirve para pasar todo a mayusculas

```python
S.upper()
```

Alpha y Digit sirve para comprobar si el vaor esta en mayuscula o minuscula

```python
print("S is alpha", S.isalpha())
print("S is digit", S.isdigit())
```

Split enumera 1 a 1 cada valor acompañado de -

```python
S = 'spams-spam-eggs-spam-bacon'
S.split('-')
```

L/Rstrip eliminan los espacios en blanco en los diferentes lados 

```python


line = 'aaa\t,bbb\t,\nccccc,\tdd\n'
print(line)

```

Rstip borra los espacios a la derecha

Lstrip borra los espacios a la izquierda
