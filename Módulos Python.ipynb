Módulos Python
==============


En este caso definimos una funcion que printa el valor de la variable [x]

```python
def printer(x):     #una función que imprime el valor de x
    print(x)        #llama a la función print para mostrar x en la consola
```

Permite llamar al directorio padre para importar modulos

```python
import sys  #importa el módulo sys para manipular rutas de sistema
sys.path.append('..') #agrega el directorio padre al path del sistema para importar módulos desde allí
```

En este caso importamos el modulo src y lo renombramos como modulo y desde el importamos la funcion printer

```python
import src.modulo as modulo      #importa el módulo llamado 'modulo' desde el directorio 'src' y lo renombra como 'modulo'
modulo.printer("Hello world!")  #llama a la función printer del módulo importado para imprimir "Hello world!"
```

En este caso importa direcatmente la funcion desde el modulo en el directorio

```python
from src.modulo import printer      #importa directamente la función printer desde el módulo 'modulo' en el directorio 'src'
printer('Hello world!')             #llama a la función printer para imprimir "Hello world!"
```

En este caso al usar el operador * import todos los modulos de los directorios

```python
from src.modulo import *       #importa todas las funciones y variables del módulo 'modulo' en el directorio 'src'
printer('Hello world!')        #llama a la función printer para imprimir "Hello world!"
```
