En este caso definimos una funcion que printa el valor de la variable [x]

```python
def printer(x):     #una función que imprime el valor de x
    print(x)        #llama a la función print para mostrar x en la consola
```

Permite llamar al directorio padre para importar modulos

```python
import sys  #importa el módulo sys para manipular rutas de sistema
sys.path.append('..') #agrega el directorio padre al path del sistema para importar módulos desde allí
```

En este caso importamos el modulo src y lo renombramos como modulo y desde el importamos la funcion printer

```python
import src.modulo as modulo      #importa el módulo llamado 'modulo' desde el directorio 'src' y lo renombra como 'modulo'
modulo.printer("Hello world!")  #llama a la función printer del módulo importado para imprimir "Hello world!"
```

En este caso importa direcatmente la funcion desde el modulo en el directorio

```python
from src.modulo import printer      #importa directamente la función printer desde el módulo 'modulo' en el directorio 'src'
printer('Hello world!')             #llama a la función printer para imprimir "Hello world!"
```

En este caso al usar el operador * import todos los modulos de los directorios

```python
from src.modulo import *       #importa todas las funciones y variables del módulo 'modulo' en el directorio 'src'
printer('Hello world!')        #llama a la función printer para imprimir "Hello world!"
```

Importa el modulo_simple como simple

```python
import src.modulo_simple as simple   #importa el módulo 'modulo_simple' desde el directorio 'src' y lo renombra como 'simple'
```

```python
simple.spam # devuelve el valor del atributo spam del módulo simple, que es 1
```

```python
simple.spam = 2                                 # cambia el valor del atributo spam del módulo simple a 2
import src.modulo_simple as simple              # vuelve a importar el módulo simple
simple.spam                                     #  devuelve el valor del atributo spam del módulo simple, que es ahora
# 2
```

```python
import importlib # importa el módulo importlib para recargar módulos
importlib.reload(simple)                     # recarga el módulo simple desde el archivo fuente
```

## Cambiando mutables en módulos ##

```python

x = 1 # asigna el valor 1 a la variable x
y = [1, 2] # asigna una lista con los valores 1 y 2 a la variable y
```

Importamos las variables desde otro modulo

```python
from src.modulo_small import x, y       # importa las variables x e y desde el módulo 'modulo_small' en el directorio 'src'
x = 42                                  #  cambia el valor de x a 42
y[0] = 42                               #   cambia el primer elemento de la lista y a 42
x, y
```

x no es un objeto compartido mutable, pero y sí:



```python
import src.modulo_small as small    # importa el módulo 'modulo_small' desde el directorio 'src' y lo renombra como 'small'
small.x                             #   devuelve el valor de la variable x del módulo small

small.x, small.y                    #  devuelve una tupla con los valores de las variables x e y del módulo small
```

## Cambios de nombres entre ficheros ##



```python
from src.modulo_small import x, y        # importa las variables x e y desde el módulo 'modulo_small' en el directorio 'src'
x = 42                                  #  cambia el valor de x a 42
import src.modulo_small as small        # importa el módulo 'modulo_small' desde el directorio 'src' y lo renombra como 'small'
x, small.x                     # devuelve una tupla con los valores de x (42) y small.x (1)
      
```

```python
importlib.reload(small)                    # recarga el módulo small desde el archivo fuente
```

## Diccionarios de espacios de nombres: dict ##

Creamos una lista con las claves de un diccionario 

```python
list(small.__dict__.keys())  # devuelve una lista con las claves del diccionario de atributos del módulo small
```

Le pedimos que filtre y solo importe las claves que empiecen por un determinado caracter

```python
list(name for name in small.__dict__.keys() if not name.startswith('__'))   # crea una lista con las claves del diccionario de atributos del módulo small que no comienzan con '__'
```
