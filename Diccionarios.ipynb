## literal syntax ##

En este primer caso primero creamos un diccionario vacio y en la segunda row creamos un diccionario con valores predeterminados

```python
d = {} 
d = {'key': 'value'}
```

En este caso hace una copia del diccionario referenciado

```python
d = {**other_dict}  # Hace una copia superficial de other_dict
```

En este caso actuaiza un diccionario con el contenido de otro 

```python
d = {**otherdict, **yet_another_dict}  # Combina dos diccionarios en uno nuevo
```

 ## dict comprehension ##

 En este caso crea un nuevo diccionario llamado d a partir de una lista de tuplas  asignando  variables en cada iteración y  desempaqueta la tupla en las variables k (clave) y v (valor).

```python
d = {k:v for k,v in [('key', 'value',)]}
```

# built-in class: dict() #

```python
d = dict() # Crea un diccionario vacío
d = dict(key='value') # Crea un diccionario con un par clave-valor
d = dict([('key', 'value')]) # Crea un diccionario a partir de una lista de tuplas
d = dict(**otherdict) # Crea un diccionario copiando otro diccionario
```

# modifying a dict #

En este caso crea una nueva clave y le añade el valor 42

```python
d['newkey'] = 42  # Agrega un nuevo par clave-valor
```

Tambien se pueden añadir listas como clave valor 

```python
d['new_list'] = [1, 2, 3] # Agrega una lista como valor
d['new_dict'] = {'nested_dict': 1}# Agrega un diccionario anidado como valor
```

Con el siguiente comando podeos eliminar pares clave-valor

```python
del d['newkey']  # Elimina el par clave-valor con clave 'newkey'
```
