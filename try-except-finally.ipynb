try /except /finally
====================

## Morfologia de try/except/finally ##





```python
try:    # Bloque de código que puede generar una excepción
    statements     # Bloque de código a ejecutar si no se produce ninguna excepción    
except name1:      # Bloque de código para manejar la excepción name1
    statements     # Bloque de código a ejecutar si se produce la excepción name1            
except (name2, name3):  # Bloque de código para manejar múltiples excepciones
    statements          # Bloque de código a ejecutar si se produce name2 o name3        
except name4 as var:    # Bloque de código para manejar la excepción name4 y asignarla a la variable var
    statements          # Bloque de código a ejecutar si se produce la excepción name4         
except:               # Bloque de código para manejar cualquier otra excepción no especificada
    statements          # Bloque de código a ejecutar si se produce cualquier otra excepción          
    statements         
    statements
else:               # Bloque de código que se ejecuta si no se produce ninguna excepción
    statements                      
```


## Ejemplos try ##


En este codigo al generar un index error a proposito el codigo dde try no se ejecutara mientras que el de fially siempre se ejecuta

```python
sep = '-' * 45 + '\n'             
print(sep + 'EXCEPTION RAISED AND CAUGHT')         
try:            
    x = 'spam'[99]    # esto generará un IndexError:
    print('except run') # no se ejecutará
finally:                   # este bloque se ejecutará siempre
    print('finally run')      
print('after run') # se ejecutará normalmente
```

En este caso como no hay ningun IndexError todo el codigo de except no se ejecutara 

```python
print(sep + 'NO EXCEPTION RAISED')
try:
    x = 'spam'[3]   # esto no generará ninguna excepción
except IndexError: # si se genera un IndexError, se manejará aquí
    print('except run')     # no se ejecutará
finally:                    # este bloque se ejecutará siempre
    print('finally run')
print('after run')
```

En este caso al no generara ningun IndexError solo se ejecutaran el codigo de try else y finally el except no se ejecutara 

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE') 
try:                # esto no generará ninguna excepción
    x = 'spam'[3]
except IndexError:    # si se genera un IndexError, se manejará aquí
    print('except run') # no se ejecutará
else:                   # este bloque se ejecutará si no se produce ninguna excepción
    print('else run')
finally:                 # este bloque se ejecutará siempre
    print('finally run')
print('after run')
```

En este caso al generar un error diferente a IndexError el except no se ejecuta y el finally si 

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT') 
try: # esto generará una excepción diferente a IndexError
    x = 1 / 0
except IndexError: # si se genera un IndexError, se manejará aquí
    print('except run') # no se ejecutará
finally: # este bloque se ejecutará siempre
    print('finally run')
print('after run')
```
