In [10]: En este caso es muy similar solo que al realizar la copia de una variable local lo que se modifica al cambiar el valor de (b) es la copia y no la varianle global.

```python
l = [1, 2]
def changer(a, b):
    b = b[:]		# Copy input list so we don't impact caller
    a = 2
    b[0] = 'spam'	# Changes our list copy only

changer(x, l)
assert l == [1, 2], "l es una invariante"
print("l = ", l) 
```

In [18]: En este caso al añadir el caracter * obligamos a que las variables que le precedan tengan que ser definidas a la hora de llamar a la funcion es decir necesitan de un valor 

```python
def kwonly(a, *b, c):
    print(a, b, c)

kwonly(1, 2, c=3)       # 1 (2,) 3

kwonly(a=1, c=3)        # 1 () 3

kwonly(1, 2, 3)         # TypeError: kwonly() missing 1 required keyword-only argument: 'c'

```

In [19]: En este caso al tener las variables obligatorias predefinidas en caso de no referenciarlas estas printaran igualmente un valor pero en caso de añadir un valor que pueda hacer referencia a estas variables obligatorias degbera especificarse a cual pertenece ese nuevo valor ya que en caso ontrario da error

```python
def kwonly(a, *, b='spam', c='ham'):
	print(a, b, c)

kwonly(1)               # 1 spam ham
kwonly(1, c=3)          # 1 spam 3
kwonly(a=1)             # 1 spam ham
kwonly(c=3, b=2, a=1)   # 1 2 3
kwonly(1, 2)            # TypeError: kwonly() takes 1 positional argument but 2 were given
```

In [2]: En este caso Args pilla todos los valores referenciados al llamar a la funcion y los guarad como una tupla es decir una lista invariable en caso de venir acompañada de (*)

```python
def f(*args):
    print(args)

f()                 # ()
f(1,)               # (1,)
f(1, 2, 3, 4)       # (1, 2, 3, 4) 
```

```python
def f(**args):
    print(args)
f()                 # {}
f(a=1, b=2)         # {'a': 1, 'b': 2}
```
